##Mock Question: Machine Learning Modelling

I want to first create a Pandas DataFrame of the uploaded dataset "loan_data(1)". I will do this using GenAI and the prompt "Please generate Python code to load a CSV file called "loan_data(1).csv". Create a pandas dataframe of this csv file and display the first 5 rows of the dataset".

```python
import pandas as pd

df = pd.read_csv('loan_data(1).csv')
display(df.head())
```

I now want to seperate the features and the label so that I can create a ML modelling model. I will use GenAI and the prompt "Please write Python code to separate the features (X) and label (y) for a classification task, where the label is "loan_status"."

```python
# Separate features (X) and label (y)
X = df.drop('loan_status', axis=1)
y = df['loan_status']

print("Shape of features (X):")
print(X.shape)
print("\nShape of label (y):")
print(y.shape)

print("\nFirst 5 rows of features (X):")
display(X.head())
print("\nFirst 5 rows of label (y):")
display(y.head())
```

I now want to scale the features, as the task described. I will use GenAI and the prompt "Please generate code to scale all numeric features using StandardScaler from scikit-learn".

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Identify numeric columns
numeric_cols = X.select_dtypes(include=['number']).columns

# Initialize StandardScaler
scaler = StandardScaler()

# Scale the numeric features
X_scaled = X.copy()
X_scaled[numeric_cols] = scaler.fit_transform(X[numeric_cols])

print("First 5 rows of scaled features (X_scaled):")
display(X_scaled.head())
```

I now want to split the dataset into training and test sets. I will use GenAI and the prompt "Please provide Python code to split the dataset into training and test sets, with 80% training and 20% test data".

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

The next step in machine learning modelling would be to train multiple ML models on the training data, and then comparing using a classification model-appropriate scoring system. I will use GenAI and the prompt "Please write Python code to train and compare multiple classification models: Logistic Regression, Random Forest, and K-Nearest Neighbours. Please evaluate using accuracy score".

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Initialize models
logistic_model = LogisticRegression(random_state=42, solver='liblinear') # Added solver for older sklearn versions
random_forest_model = RandomForestClassifier(random_state=42)
k_neighbors_model = KNeighborsClassifier()

models = {
    'Logistic Regression': logistic_model,
    'Random Forest': random_forest_model,
    'K-Nearest Neighbors': k_neighbors_model
}

results = {}

for name, model in models.items():
    print(f"\nTraining {name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy
    print(f"{name} Accuracy: {accuracy:.4f}")

print("\n--- Model Comparison ---")
for name, accuracy in results.items():
    print(f"{name}: {accuracy:.4f}")
```

And finally, for a more visual way of comparing the ML models, we can generate a barchart to compare the accuracy score of each of the models. I will use GenAI and the prompt "Please provide Python code to create a bar chart comparing the accuracy score of the models".

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Convert results dictionary to a pandas DataFrame for easy plotting
results_df = pd.DataFrame(list(results.items()), columns=['Model', 'Accuracy'])

# Create the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Accuracy', data=results_df, palette='viridis', hue='Model', legend=False)
plt.title('Comparison of Model Accuracy Scores')
plt.xlabel('Machine Learning Model')
plt.ylabel('Accuracy Score')
plt.ylim(0, 1) # Accuracy scores range from 0 to 1
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

From the accuracy scores, coupled with the barchart, we can see that the Random Forest model is performing the best, followed by KNN, and then Logistic Regression.

The Random Forest model is perhaps the best as it is efficiently capturing the non-linear pattern in the data and between features, whereas the KNN model also has a relatively strong performance, but perhaps does not as effectively capture the features. Logistic regression has a strong baseline, but again, the Random Forest model is likely a little more complex and therefore performs better in relation to the dataset.

The next step would be to try to further improve the model via hyperparameter tuning:

Let's first improve the model via hyperparameter tuning. I will use GenAI along with the prompt: "Please write Python code to perform hyperparameter tuning for Logistic Regression, Random Forest, and K-Nearest Neighbors using GridSearchCV. Use the following parameters: Logistic Regression (C, solver), Random Forest (n_estimators, max_depth, min_samples_split), KNN (n_neighbors, metric). Print the best parameters and accuracy for each model."

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Define models and their parameter grids
models_and_params = {
    'Logistic Regression': {
        'model': LogisticRegression(random_state=42, solver='liblinear'),
        'params': {
            'C': [0.1, 1, 10],
            'solver': ['liblinear']
        }
    },
    'Random Forest': {
        'model': RandomForestClassifier(random_state=42),
        'params': {
            'n_estimators': [50, 100, 200],
            'max_depth': [None, 10, 20],
            'min_samples_split': [2, 5]
        }
    },
    'K-Nearest Neighbors': {
        'model': KNeighborsClassifier(),
        'params': {
            'n_neighbors': [3, 5, 7],
            'metric': ['euclidean', 'manhattan']
        }
    }
}

best_models_results = {}

for name, mp in models_and_params.items():
    print(f"\n--- Performing GridSearchCV for {name} ---")
    grid_search = GridSearchCV(
        estimator=mp['model'],
        param_grid=mp['params'],
        cv=5, # 5-fold cross-validation
        scoring='accuracy',
        n_jobs=-1, # Use all available cores
        verbose=1
    )

    grid_search.fit(X_train, y_train)

    best_models_results[name] = {
        'best_params': grid_search.best_params_,
        'best_accuracy': grid_search.best_score_
    }

    print(f"Best parameters for {name}: {grid_search.best_params_}")
    print(f"Best accuracy for {name}: {grid_search.best_score_:.4f}")

print("\n--- Hyperparameter Tuning Results ---")
for name, results in best_models_results.items():
    print(f"{name}:\n  Best Params: {results['best_params']}\n  Best Accuracy: {results['best_accuracy']:.4f}")
```

This was taking a bit too long to run on my GPU, so we'll be trying a different approach. I want to try to apply cross-validation on my models for a comparison measure. I will use GenAI with the prompt: "Please write Python code to evaluate Logistic Regression, Random Forest, and KNN using 5-fold cross-validation. Please print the mean accuracy for each model".

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Initialize models (using the same configurations as before)
logistic_model = LogisticRegression(random_state=42, solver='liblinear')
random_forest_model = RandomForestClassifier(random_state=42)
k_neighbors_model = KNeighborsClassifier()

models = {
    'Logistic Regression': logistic_model,
    'Random Forest': random_forest_model,
    'K-Nearest Neighbors': k_neighbors_model
}

print("--- 5-Fold Cross-Validation Mean Accuracy ---")
for name, model in models.items():
    print(f"\nEvaluating {name}...")
    # Perform 5-fold cross-validation
    scores = cross_val_score(model, X_scaled, y, cv=5, scoring='accuracy', n_jobs=-1)
    mean_accuracy = scores.mean()
    print(f"{name} Mean Accuracy (5-fold CV): {mean_accuracy:.4f}")
```
